```python
def extract_names(mat_arr):
    names = {}
    flat = mat_arr.flatten()

    for i, cell in enumerate(flat):
        true_id = i + 1  

        if cell.size == 0:
            continue

        value = cell
        if hasattr(value, "__len__"):
            value = value[0]

        if hasattr(value, "__len__") and not isinstance(value, str):
            value = value[0]

        if isinstance(value, str):
            names[true_id] = value.strip()

    return names
```

```python
from tqdm import tqdm

image_root="../image"
dataset_root="../dataset"
train_split="../train_test_split/classification/train.txt"
copied=0
skipped=0
```

```python
with open(train_split,"r") as f:
    lines=f.readlines()

for line in tqdm(lines):
    rel_path=line.strip()
    if not rel_path:
        continue
    
    src_path=os.path.join(image_root,rel_path)
    if not os.path.exists(src_path):
        skipped+=1
        continue
    
    try:
        brand_id,model_id, _, _=rel_path.split("/")
        brand_id=int(brand_id)
        model_id=int(model_id)
    except Exception:
        skipped+=1
        continue
    
    if brand_id not in brand_names or model_id not in model_names:
        skipped+=1
        continue
    
    brand=normalize_name(brand_names[brand_id])
    model=normalize_name(model_names[model_id])
    class_name=f"{brand}_{model}".replace(" ","_")
    
    if class_name not in valid_classes:
        skipped+=1
        continue
    
    dst_dir=os.path.join(dataset_root,"train",class_name)
    os.makedirs(dst_dir,exist_ok=True)
    
    dst_path=os.path.join(dst_dir,os.path.basename(src_path))
    shutil.copy2(src_path,dst_path)
    copied+=1
    
print("copied:",copied,"skipped:",skipped)
    
```

creating validation split

```python
val_dir=os.path.join(dataset_root,"val")
os.makedirs(val_dir,exist_ok=True)
```

```python
import random
```

```python
val_ratio=0.2
train_dir=os.path.join(dataset_root,"train")
val_dir=os.path.join(dataset_root,"val")
moved=0
unmoved=0

for cls in os.listdir(train_dir):
    cls_train_path=os.path.join(train_dir,cls)
    if not os.path.isdir(cls_train_path):
        unmoved+=1
        continue
    
    images=os.listdir(cls_train_path)
    random.shuffle(images)
    
    val_count=int(len(images)*val_ratio)
    if val_count==0:
        unmoved+=1
        continue
    
    cls_val_path=os.path.join(val_dir,cls)
    os.makedirs(cls_val_path,exist_ok=True)
    
    for img in images[:val_count]:
        src=os.path.join(cls_train_path,img)
        dst=os.path.join(cls_val_path,img)
        shutil.move(src,dst)
        moved+=1
        
print("moved:",moved)
print("unmoved:",unmoved)
```
